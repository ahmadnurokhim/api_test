```python
import numpy as np
data = [{'timestamp': 0, 'temperature': 1, 'humidity': 1}]
for x in range(100):
    data.append({
        'timestamp': data[-1]['timestamp']+1,
        'temperature': (data[-1]['temperature'] * np.random.normal(1, 0.1)),
        'humidity': (data[-1]['humidity'] * np.random.normal(1, 0.1)),
    })
for x in data:
    print(f"{x},")
```
